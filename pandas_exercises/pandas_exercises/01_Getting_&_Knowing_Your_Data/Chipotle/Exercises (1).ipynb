### Step 13. Turn the item price into a float

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

#### Step 13.a. Check the item price type

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(
    lambda x: float(x.replace('$', '')) if isinstance(x, str) else x
)

```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Total revenue:", revenue)
```

### Step 15. How many orders were made in the period?





```python
num_orders = chipo['order_id'].nunique()
print("Number of unique orders:", num_orders)
```

### Step 16. What is the average revenue amount per order?

```python
# 1. total revenue
total_revenue = (chipo['quantity'] * chipo['item_price']).sum()

# 2. total orders
total_orders = chipo['order_id'].nunique()

# 3. average revenue per order
average_revenue_per_order = total_revenue / total_orders

print("Average revenue per order:", round(average_revenue_per_order, 2))

```
