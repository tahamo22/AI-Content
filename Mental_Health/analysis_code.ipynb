```python
def kpi_distribution_by_occupation(column_name):
    # Clean values
    df[column_name] = df[column_name].astype(str).str.strip().str.capitalize()

    # Unique Values in the Column
    unique_values = df[column_name].unique()

    # Loop over each unique value in the column (e.g., High, Medium, Low)
    for val in unique_values:
        print(f"\nValue: {val}")

        # Students
        students_total = df[df['Occupation'] == 'Student'].shape[0]
        students_val = df[(df['Occupation'] == 'Student') & (df[column_name] == val)].shape[0]
        students_percentage = (students_val / students_total) * 100 if students_total > 0 else 0

        # Self-employed
        self_employed_total = df[df['self_employed'] == 'Yes'].shape[0]
        self_employed_val = df[(df['self_employed'] == 'Yes') & (df[column_name] == val)].shape[0]
        self_employed_percentage = (self_employed_val / self_employed_total) * 100 if self_employed_total > 0 else 0

        # Employed
        employed_total = df[df['self_employed'] == 'No'].shape[0]
        employed_val = df[(df['self_employed'] == 'No') & (df[column_name] == val)].shape[0]
        employed_percentage = (employed_val / employed_total) * 100 if employed_total > 0 else 0

        # Print results
        print(f"Students: {students_percentage:.2f}% | Self-employed: {self_employed_percentage:.2f}% | Employed: {employed_percentage:.2f}%")

# Example Usage:
kpi_distribution_by_occupation('Days_Indoors')
```

```python
def family_history_vs_mood_swings():
    occupations = ['Student', 'Self-employed', 'Employed']
    results = {'Occupation': [], 'Family History & High Mood Swings (%)': []}

    for occ in occupations:
        if occ == 'Employed':
            occ_filter = (df['self_employed'] == 'No')
        else:
            occ_filter = (df['Occupation'] == occ)

        total_occ = df[occ_filter].shape[0]
        family_and_high_mood = df[(occ_filter) & (df['family_history'] == 'Yes') & (df['Mood_Swings'] == 'High')].shape[0]

        percentage = (family_and_high_mood / total_occ) * 100 if total_occ > 0 else 0
        results['Occupation'].append(occ)
        results['Family History & High Mood Swings (%)'].append(percentage)

    # Visualization
    plt.figure(figsize=(10,6))
    plt.bar(results['Occupation'], results['Family History & High Mood Swings (%)'])
    plt.title('Impact of Family History on Severe Mood Swings', fontsize=16)
    plt.ylabel('Percentage (%)', fontsize=14)
    plt.ylim(0, 100)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Run KPI Visualization
family_history_vs_mood_swings()
```
