```python
import requests
import base64
import re
import os


save_path = "E:\conda\images\phone_image.jpg"


save_dir = os.path.dirname(save_path) or "."
os.makedirs(save_dir, exist_ok=True)


image = MyWebsite.find("div", class_="img-c")
if image:
    image_url = image.find("img")["src"]
    
    # التحقق مما إذا كان الرابط هو Base64
    if image_url.startswith("data:image"):
        base64_string = re.search(r"base64,(.*)", image_url).group(1)
        image_data = base64.b64decode(base64_string)
        with open(save_path, "wb") as file:
            file.write(image_data)
        print(f"Image downloaded successfully (Base64) to {save_path}!")
    else:
        try:
            response = requests.get(image_url)
            response.raise_for_status()
            with open(save_path, "wb") as file:
                file.write(response.content)
            print(f"Image downloaded successfully (HTTP) to {save_path}!")
        except requests.exceptions.RequestException as e:
            print("Error downloading image:", e)
else:
    print("Image div not found!")
```
